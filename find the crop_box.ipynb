```python
%pip install opencv-python
import cv2
import SimpleITK as sitk
import numpy as np

img = sitk.ReadImage(r"E:\CMU courses\2026 Spring\16725 Medical Image Analysis\final project\data\US_labeled\Extracted\URS08_H1\URS08_H1\Labels\0.mhd")
img_np = sitk.GetArrayFromImage(img)


img_show = img_np.copy()


img_show = img_show.astype(np.float32)
img_show = (img_show - img_show.min()) / (img_show.max() - img_show.min() + 1e-8)
img_show = (img_show * 255).astype(np.uint8)

roi = cv2.selectROI("Select ROI", img_show, showCrosshair=True, fromCenter=False)
cv2.destroyAllWindows()

x, y, w, h = roi
print("ROI =", roi)
print("crop_box =", (x, y, x + w, y + h))
```
